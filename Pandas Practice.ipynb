```python
def load_file(filename):
    with open(filename, encoding="utf-8-sig") as f:
        data, cols = [], []
        for i, line in enumerate(f.read().splitlines()):
            if i == 0:
                cols += line.split(",")
            else:
                data.append([float(x) for x in line.split(",")])
        df = pd.DataFrame(data, columns=cols)
    return df
load_file(filename).head()
```

```python
# Some valid and invalid email address
import re
text = """Mavia maviaali@gmail.com
Ziyad ziyad_beyg@yahoo.com
Aniq aniqsaudgmail.com
Salman salman@gmailcom
Sherry sherry@gmail.pk
Hammad hammad@gmail.c
Saif saif@gmail.compk
"""
```

```python
# creating pattern to filtering out the correct email address
pattern = '[a-z0-9._]+[@]\w+[.]\w{2,3}' #text@text.com
```

```python
regex = re.compile(pattern, flags=re.IGNORECASE)
regex
```

```python
regex.findall(text)
```
